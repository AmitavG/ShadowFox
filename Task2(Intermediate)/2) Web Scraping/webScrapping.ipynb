```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/UEFA_Champions_League'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
soup.find('table')
```

```python
soup.find_all('table')
```

```python
soup.find_all('table')[9]
```

```python
soup.find('table', class_='wikitable plainrowheaders sortable')
```

```python
tables = soup.find_all('table')[9]
```

```python
print(tables)
```

```python
nation_titles= tables.find_all('th')
```

```python
nation_titles = nation_titles[:4]
nation_titles
```

```python
nation_table_titles = [title.text.strip() for title in nation_titles]

print(nation_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = nation_table_titles)
df
```

```python
column_data = tables.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all(['th', 'td'])
    nation_table_data = [data.text.strip() for data in row_data[:4]]

    df.loc[len(df)] = nation_table_data
```
